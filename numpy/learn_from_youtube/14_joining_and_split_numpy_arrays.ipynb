Join & Split Functions NumPy Arrays

```python
import numpy as np

var = np.array([1,2,3,4])
var1 = np.array([9,8,7,6])

ar = np.concatenate((var, var1))
print(ar)
```

```python
vr = np.array([[1,2],[3,4]])
vr1 = np.array([[9,8],[7,6]])

ar = np.concatenate((vr, vr1), axis=1) # row-wise axis=1

print(vr)
print()
print(vr1)
print()
print(ar)
```

```python
vr = np.array([[1,2],[3,4]])
vr1 = np.array([[9,8],[7,6]])

ar = np.concatenate((vr, vr1), axis=0) # column-wise axis=0

print(vr)
print()
print(vr1)
print()
print(ar)
```

```python
var_1 = np.array([[1,2],[3,4]])
var_2 = np.array([[9,8],[7,6]])

ar = np.stack((var_1, var_2), axis=1) 

print(var_1)
print()
print(var_2)
print()
print(ar)
```

```python
var_1 = np.array([1,2,3,4])
var_2 = np.array([9,8,7,6])

ar = np.stack((var_1, var_2)) 

print(var_1)
print()
print(var_2)
print()
print(ar)
```

```python
var_1 = np.array([1,2,3,4])
var_2 = np.array([9,8,7,6])

ar = np.hstack((var_1, var_2)) 

print(var_1)
print()
print(var_2)
print()
print(ar)
```

```python
var_1 = np.array([1,2,3,4])
var_2 = np.array([9,8,7,6])

ar = np.vstack((var_1, var_2)) 

print(var_1)
print()
print(var_2)
print()
print(ar)
```

```python
var_1 = np.array([1,2,3,4])
var_2 = np.array([9,8,7,6])

ar = np.dstack((var_1, var_2)) 

print(var_1)
print()
print(var_2)
print()
print(ar)
```

```python
var_1 = np.array([1,2,3,4])
var_2 = np.array([9,8,7,6])

ar = np.stack((var_1, var_2), axis=1)
ar1 = np.hstack((var_1, var_2)) # row
ar2 = np.vstack((var_1, var_2)) # columns
ar3 = np.dstack((var_1, var_2)) # height

 

print(ar)
print()
print(ar1)
print()
print(ar2)
print()
print(ar3)
```

```python
var = np.array([1,2,3,4,5,6])

print(var)

ar = np.array_split(var, 3)
print()
print(ar)
```
