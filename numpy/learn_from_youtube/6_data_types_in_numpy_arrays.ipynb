Data Types in NumPy Arrays:
- 


```python
import numpy as np

var = np.array([1,2,3,4])

print("Data type: ",var.dtype)
```

```python
var = np.array([1.0,1.2,1.3,4.2])

print("Data type: ",var.dtype)
```

```python
var = np.array(["a","s","F","D"])

print("Data type: ",var.dtype)
```

```python
var = np.array(["A","p",1,3,6])

print("Data type: ",var.dtype)
```

```python
x = np.array([1,2,3,4])

print("Data type: ",x.dtype)
```

```python
x = np.array([1,2,3,4], dtype = np.int8) #int 32 to int8

print("Data type: ", x.dtype)
print(x)
```

```python
x1 = np.array([1,2,3,4], dtype = "f") #int 32 to float

print("Data type: ", x1.dtype)
print(x1)
```

```python
x2 = np.array([1,2,3,4])

new = np.float32(x2)

print("Data type: ", x2.dtype)
print(x2)
print(new)
print("Data type of new: ", new.dtype)
```

```python
new_one = np.int_(new)

print(new_one)
print("Data type: ", new_one.dtype)
```

```python
x3 = np.array([1,2,3,4])

new_1 = x3.astype(float)
print(x3)
print(new_1)
```
