Arithmetic Functions in NumPy Arrays:
- 

```python
import numpy as np

var = np.array([3,6,9,2,44])

print("Minimum: ", np.min(var))
print("Maximum: ", np.max(var))
```

```python
# to know location of min use argmin, for max use argmax

print("Minimum: ", np.min(var), "Index: ", np.argmin(var))
print("Maximum: ", np.max(var), "Index: ", np.argmax(var))
```

```python
var1 = np.array([[2,1,3], [9,5,6]])

print(np.min(var1, axis=1))

# axis = 0 -> column
# axis = 1 -> row
```

```python
var1 = np.array([[2,1,3], [9,5,6]])

print(np.min(var1, axis=0))

# axis = 0 -> column
# axis = 1 -> row
```

```python
var = np.array([9, 4, 36])

print("sqrt: ", np.sqrt(var))
```

```python
var2 = np.array([1,2,3])

print(np.sin(var2))
print(np.cos(var2))
```
