Arithmetic Operations in NumPy Arrays:
- 

```python
import numpy as np

var = np.array([1,2,3,4])

varadd = var + 3

print(varadd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

# varadd = var1 + var2

varadd = np.add(var1, var2)

print(varadd)
```

```python
var = np.array([1,2,3,4])

varadd = var - 3

print(varadd)
```

```python
var = np.array([1,2,3,4])

varadd = var * 3

print(varadd)
```

```python
var = np.array([1,2,3,4])

varadd = var / 3

print(varadd)
```

```python
var = np.array([1,2,3,4])

varadd = var % 3

print(varadd)
```

```python
# two dimensional arrays

var21 = np.array([[1,2,3,4],[1,2,3,4]])
var22 = np.array([[1,2,3,4],[1,2,3,4]])

varadd2 = var21 * var22
print(varadd2)
```

```python
# reciprocal

var = np.array([1,2,3,4])

varnew = np.reciprocal(var)

print(varnew)
```
