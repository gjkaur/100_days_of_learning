Indexing & Slicing in NumPy Arrays:



```python
import numpy as np

var = np.array([9,8,7,6])

print(var[1])
print(var[-3])
```

```python
var1 = np.array([[9,8,7],[4,5,6]])

print(var1)
print(var1.ndim)
print()

print(var1[0,1])
print(var1[0,2])
```

```python
var2 = np.array([[[1,2],[6,7]]])

print(var2)
print(var2.ndim)

print()
print(var2[0,1,1])
```

```python
import numpy as np

var = np.array([1,2,3,4,5,6,7])

print(var)
```

```python
print("2 to 5: ", var[1:5]) # excludes last index number
# index starts from zero
```

```python
print("2 to end: ", var[1:])
```

```python
print("start to 5: ", var[:5])
```

```python
print("stop: ", var[::])
```

```python
print("stop: ", var[::2]) # step=2
```

```python
print("stop: ", var[1:6:2])
```

```python
var1 = np.array([[1,2,3,4,5],[9,8,7,6,5],[11,12,13,14,15]])

print(var1)
```

```python
print("8 to 5: ", var1[1,1:])
```
